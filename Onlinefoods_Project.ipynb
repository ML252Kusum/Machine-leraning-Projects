```python
## Now handling Outliers present in data

num_vars=df3.select_dtypes(include=['int','float'])
num_vars
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(35,35))
for i,var in enumerate(num_vars):
    plt.subplot(7,6,i+1)
    sns.boxplot(num_vars[var])
```

```python
pd.DataFrame(num_vars,columns=['Age'])
```

```python
#Removing outliers of Age Feature

sns.boxplot(num_vars['Age'])
```

```python
Q1=num_vars['Age'].quantile(.25)
print("Q1:",Q1)
Q3=num_vars['Age'].quantile(.75)
print("Q3:",Q3)
IQR=Q3-Q1
print("IQR:",IQR)
```

```python
ll=Q1-(1.5*IQR)
ul=Q3+(1.5*IQR)
ll,ul
```

```python
# Now we got lower and upper limit so we just need to remove all the values lower than 18.5 and higher than 30.5
num_vars=num_vars[(num_vars['Age']>ll)&(num_vars['Age']<ul)] 
```

```python
sns.boxplot(num_vars['Age'])  ## Hence outliers of age Coulmn get removed.
```
